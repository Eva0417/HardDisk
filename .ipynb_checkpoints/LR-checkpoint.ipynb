```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import validation_curve
import numpy as np
import pandas as pd
from sklearn.metrics import recall_score,accuracy_score,precision_score,auc,make_scorer
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
import os
from sklearn.model_selection import learning_curve
```

```python
real_data = shuffle(pd.read_csv('./data_Q1_2017/2017-01-01.csv'))
col = ['smart_12_raw','smart_197_raw','smart_187_raw','smart_4_raw','smart_5_raw','smart_189_raw','smart_183_raw','smart_199_raw','smart_198_raw','smart_188_raw','smart_194_raw','smart_190_raw','smart_3_raw','smart_184_raw','smart_192_raw','smart_191_raw']
file_dir = "./data_Q1_2017/"  # file directory
all_csv_list = os.listdir(file_dir)  # get csv list
embedding_dim = 16
seq_length = 60
num_classes = 2
keep_prob = 0.5
learning_rate = 1e-4
batch_size = 256
num_epochs = 5
lstm_size = 3 * 16
falls = pd.Series(np.ones(100))
```

```python
maked_fake = np.load('./maked_fake.npy')
restore = pd.read_csv('Failure_16.csv',index_col=[0])
restore = restore.max()
for i in range(maked_fake.shape[0]):
    for j in range(maked_fake.shape[1]):
        maked_fake[i][j] = restore[j]*(maked_fake[i][j] + 1)/2
maked_fake = pd.DataFrame(columns=col, data=maked_fake)
```

```python
X_train_RL = np.array([]).reshape(0,embedding_dim)
Y_train_RL = np.array([]).reshape(0,2)
for i in range(10):
    newframe = pd.DataFrame(columns=real_data.columns)
    for single_csv in all_csv_list:
        if single_csv.find(".csv", 0, len(single_csv)) != -1:
            print(os.path.join(file_dir, single_csv))
            data = pd.read_csv(os.path.join(file_dir, single_csv)).fillna(0)
            newframe = newframe.append(data[data['serial_number'] == real_data.iloc[i].serial_number],ignore_index=True)
    
    y_frame = newframe.sort_values(by='date')['failure']
    x_frame = newframe.sort_values(by='date')[col]
    y_frame = pd.concat([y_frame,falls],axis=0,ignore_index=True)
    x_frame = x_frame.append(maked_fake.iloc[i*100:i*100+100])
    temp = pd.Series(1 - y_frame.values)
    y = pd.concat([y_frame,temp], axis=1,ignore_index=True)
    x_train = np.array(x_frame)
    y_train = np.array(y)
    X_train_RL = np.concatenate((X_train_RL,x_train))
    Y_train_RL = np.concatenate((Y_train_RL,y_train))
    print(i)
```

```python
for i in range(3):
    X_test_RL = np.array([]).reshape(0,embedding_dim)
    Y_test_RL = np.array([]).reshape(0,2)
    for i in range(1):
        newframe = pd.DataFrame(columns=real_data.columns)
        for single_csv in all_csv_list:
            if single_csv.find(".csv", 0, len(single_csv)) != -1:
                print(os.path.join(file_dir, single_csv))
                data = pd.read_csv(os.path.join(file_dir, single_csv)).fillna(0)
                newframe = newframe.append(data[data['serial_number'] == real_data.iloc[i+70].serial_number],ignore_index=True)
    
        y_frame = newframe.sort_values(by='date')['failure']
        x_frame = newframe.sort_values(by='date')[col]
        temp = pd.Series(1 - y_frame.values)
        y = pd.concat([y_frame,temp], axis=1,ignore_index=True)
        x_train = np.array(x_frame)
        y_train = np.array(y)
        X_test_RL = np.concatenate((X_test_RL,x_train))
        Y_test_RL = np.concatenate((Y_test_RL,y_train))
```

```python
Sd = StandardScaler()
```

```python
LR = LogisticRegression()
```

```python
X_train_RL = X_train_RL.astype(np.float32)
Y_train_RL = Y_train_RL.astype(np.float32)
X_test_RL = X_test_RL.astype(np.float32)
Y_test_RL = Y_test_RL.astype(np.float32)
```

```python
data_0 = pd.read_csv('./data_Q1_2017/2017-12-21.csv')
data = data_0.fillna(0)
data_lable = np.array(data_0['failure'])
data = np.array(data[col])
```

```python
data = Sd.fit_transform(data)
X_train_RL_fit = Sd.fit_transform(X_train_RL)
```

```python
train_sizes,train_scores,test_scores = learning_curve(LR,X=data,y=data_lable,scoring=make_scorer(precision_score))
fig,ax=plt.subplots(1,1,figsize=(6,6))
ax.set_ylim((0.0,1.1)) 
ax.set_xlabel("training examples") # x
ax.set_ylabel("LR-auc")
ax.grid() # 
ax.plot(train_sizes,np.mean(train_scores,axis=1),'o-',color='r',label='train score')
# 
ax.plot(train_sizes,np.mean(test_scores,axis=1),'o-',color='g',label='test score')
ax.legend(loc='best') # 

plt.show()
```
