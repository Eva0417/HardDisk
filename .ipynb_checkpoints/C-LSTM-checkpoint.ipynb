```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = '1'
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, Embedding, Convolution1D, Reshape
from tensorflow.keras.optimizers import Adam
import numpy as np
```

```python
embedding_dim = 16
seq_length = 60
num_classes = 2
keep_prob = 0.5
learning_rate = 8e-4
batch_size = 256
num_epochs = 5
lstm_size = 3 * 16
filter_length = 5
nb_filter = 64  
opt = Adam(lr = learning_rate)
```

```python
model = Sequential()
model.add(Convolution1D(16,
                        kernel_size = 1,
                        activation='relu',input_shape = (seq_length,embedding_dim)))
model.add(Dense(units = embedding_dim))
model.add(LSTM(units=lstm_size,return_sequences=True,input_shape=(seq_length,embedding_dim)))
model.add(Dropout(keep_prob))
model.add(LSTM(units=lstm_size,input_shape=(seq_length,embedding_dim)))
model.add(Dropout(keep_prob))
model.add(Dense(units=num_classes,activation='softmax'))
model.compile(loss = 'binary_crossentropy', optimizer=opt,metrics=['accuracy','AUC','Precision','Recall'])
model.summary()
```

```python
X_train = np.load('./X_train.npy')
Y_train = np.load('./Y_train.npy')
```

```python
X_train.shape
```

```python
history = model.fit(X_train,Y_train, batch_size, num_epochs)
```

```python
import matplotlib.pyplot as plt
acc = history.history['accuracy']  #
loss = history.history['loss']   #
AUC = history.history['AUC']
Precision = history.history['Precision']
Recall = history.history['Recall']
epochs = range(1,len(acc)+1)

plt.plot(epochs,acc,'bo',label='C-LSTM Trainning acc',color='r',marker='o',linestyle='-.',alpha=0.5)  #epochs
plt.xlabel('epochs') 
plt.ylabel('Accuracy') 
plt.legend() #
  
plt.figure() #
plt.plot(epochs,loss,'bo',label='C-LSTM Trainning loss',color='r',marker='o',linestyle='-.',alpha=0.5)
plt.xlabel('epochs') 
plt.ylabel('Loss')
plt.legend() ##

plt.figure() #
plt.plot(epochs,AUC,'bo',label='C-LSTM Trainning AUC',color='r',marker='o',linestyle='-.',alpha=0.5)
plt.xlabel('epochs') 
plt.ylabel('AUC')
plt.legend() ##

plt.figure() #
plt.plot(epochs,Precision,'bo',label='C-LSTM Trainning Precision',color='r',marker='o',linestyle='-.',alpha=0.5)
plt.xlabel('epochs') 
plt.ylabel('Precision')
plt.legend() ##

plt.figure() #
plt.plot(epochs,Recall,'bo',label='C-LSTM Trainning Recall',color='r',marker='o',linestyle='-.',alpha=0.5)
plt.xlabel('epochs') 
plt.ylabel('Recall')
plt.legend() ##
```
