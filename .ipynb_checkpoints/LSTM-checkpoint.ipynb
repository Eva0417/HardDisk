```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = '1'
```

```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, Embedding
from tensorflow.keras.optimizers import Adam
```

```python
embedding_dim = 16
seq_length = 60
num_classes = 2
keep_prob = 0.5
learning_rate = 1e-4
batch_size = 256
num_epochs = 5
lstm_size = 3 * 16
opt = Adam(lr = learning_rate)
model = Sequential()
model.add(LSTM(units=lstm_size,return_sequences=True,input_shape=(seq_length,embedding_dim)))
model.add(Dropout(keep_prob))
model.add(LSTM(units=lstm_size,input_shape=(seq_length,embedding_dim)))
model.add(Dropout(keep_prob))
model.add(Dense(units=num_classes,activation='softmax'))
model.compile(loss = 'binary_crossentropy', optimizer=opt,metrics=['accuracy','AUC','Precision','Recall'])
model.summary()
```

# 数据准备

## 滑动窗口

```python
def build_train(train, y, n_in, n_out):
    # train = train.drop(["BillingDate"], axis=1)
    X_train, Y_train = [], []
    for i in range(train.shape[0] - n_in - n_out + 1):
        X_train.append(np.array(train.iloc[i:i + n_in]))
        #Y_train.append(np.array(y.iloc[i + n_in:i + n_in + n_out]))
        Y_train.append(np.array(y.iloc[i + n_in]))

    return np.array(X_train), np.array(Y_train)
```

## 恢复生成数据

```python
maked_fake = np.load('./maked_fake.npy')
restore = pd.read_csv('Failure_16.csv',index_col=[0])
restore = restore.max()
for i in range(maked_fake.shape[0]):
    for j in range(maked_fake.shape[1]):
        maked_fake[i][j] = restore[j]*(maked_fake[i][j] + 1)/2
maked_fake = pd.DataFrame(columns=col, data=maked_fake)
```

## 提取数据

```python
from sklearn.utils import shuffle
real_data = shuffle(pd.read_csv('./data_Q1_2017/2017-01-01.csv'))
```

```python
col = ['smart_12_raw','smart_197_raw','smart_187_raw','smart_4_raw','smart_5_raw','smart_189_raw','smart_183_raw','smart_199_raw','smart_198_raw','smart_188_raw','smart_194_raw','smart_190_raw','smart_3_raw','smart_184_raw','smart_192_raw','smart_191_raw']
falls = pd.Series(np.ones(100))
X_train = np.array([]).reshape(0,seq_length,embedding_dim)
Y_train = np.array([]).reshape(0,2)
file_dir = "./data_Q1_2017/"  # file directory
all_csv_list = os.listdir(file_dir)  # get csv list
```

```python
for i in range(10):
    newframe = pd.DataFrame(columns=real_data.columns)
    for single_csv in all_csv_list:
        if single_csv.find(".csv", 0, len(single_csv)) != -1:
            print(os.path.join(file_dir, single_csv))
            data = pd.read_csv(os.path.join(file_dir, single_csv)).fillna(0)
            newframe = newframe.append(data[data['serial_number'] == real_data.iloc[i].serial_number],ignore_index=True)
    
    y_frame = newframe.sort_values(by='date')['failure']
    x_frame = newframe.sort_values(by='date')[col]
    y_frame = pd.concat([y_frame,falls],axis=0,ignore_index=True)
    x_frame = x_frame.append(maked_fake.iloc[i*100:i*100+100])
    temp = pd.Series(1 - y_frame.values)
    y = pd.concat([y_frame,temp], axis=1,ignore_index=True)
    x_train, y_train = build_train(x_frame,y,seq_length,1)
    X_train = np.concatenate((X_train,x_train))
    Y_train = np.concatenate((Y_train,y_train))
    print(i)
```

```python
np.save('./X_train',X_train)
np.save('./Y_train',Y_train)
```

```python
X_train = X_train.astype(np.float32)
Y_train = Y_train.astype(np.float32)
```

```python
history = model.fit(X_train, Y_train, batch_size, num_epochs)
```
